```python
# =========================
# CELL: Full LangGraph Workflow (NO OLLAMA)
# Text -> Image (SDXL) -> Image -> Text (BLIP) -> Summary
# =========================

import os
from typing import TypedDict, List

import torch
from diffusers import DiffusionPipeline

from transformers import pipeline as hf_pipeline
from transformers import AutoProcessor, AutoModelForVision2Seq

from langgraph.graph import StateGraph, START, END

from IPython.display import display, Markdown
from PIL import Image as PILImage


# -------------------------
# CONFIG
# -------------------------
OUTPUT_DIR = "outputs"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Choose device
# Colab GPU: "cuda"
# Colab CPU: "cpu" (very slow for SDXL)
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"


# -------------------------
# LOAD MODELS ONCE
# -------------------------
# Text-to-Image (SDXL)
# Using float16 on GPU; fallback to bfloat16/cpu if needed
if DEVICE == "cuda":
    t2i_pipe = DiffusionPipeline.from_pretrained(
        "stabilityai/stable-diffusion-xl-base-1.0",
        torch_dtype=torch.float16
    ).to("cuda")
else:
    t2i_pipe = DiffusionPipeline.from_pretrained(
        "stabilityai/stable-diffusion-xl-base-1.0",
        torch_dtype=torch.bfloat16,
        device_map="cpu"
    )

t2i_pipe.enable_attention_slicing()

# Image-to-Text (BLIP)
# You gave both pipeline and processor/model — we incorporate both.
i2t_pipe = hf_pipeline("image-to-text", model="Salesforce/blip-image-captioning-base", device=0 if DEVICE == "cuda" else -1)

processor = AutoProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
vision_model = AutoModelForVision2Seq.from_pretrained("Salesforce/blip-image-captioning-base")
# (We keep processor/model loaded for completeness / later control; pipeline is used for inference.)


# -------------------------
# STATE
# -------------------------
class WFState(TypedDict, total=False):
    prompt: str
    image_path: str
    caption: str
    summary: str


# -------------------------
# CHECKPOINT helper (shows which step is executing)
# -------------------------
def checkpoint(msg: str):
    display(Markdown(f"✅ **Checkpoint:** {msg}"))


# -------------------------
# NODES
# -------------------------
def generate_image(state: WFState) -> WFState:
    prompt = state["prompt"]
    out_path = os.path.join(OUTPUT_DIR, "generated.png")

    checkpoint("Generating image with SDXL (Diffusers)")
    img = t2i_pipe(prompt).images[0]
    img.save(out_path)

    return {"image_path": out_path}


def caption_image(state: WFState) -> WFState:
    image_path = state["image_path"]

    checkpoint("Generating caption with BLIP (Transformers image-to-text)")
    result = i2t_pipe(image_path)  # typically [{"generated_text": "..."}]
    caption = ""
    if isinstance(result, list) and len(result) > 0:
        caption = result[0].get("generated_text", "").strip()
    else:
        caption = str(result).strip()

    return {"caption": caption}


def build_summary(state: WFState) -> WFState:
    prompt = state["prompt"]
    caption = state.get("caption", "")

    checkpoint("Building final summary (rule-based, no LLM)")
    summary = (
        f"- **Prompt:** {prompt}\n"
        f"- **Caption (BLIP):** {caption}\n"
        f"- **Output file:** {state.get('image_path','')}\n"
    )
    return {"summary": summary}


# -------------------------
# BUILD GRAPH (Sequential)
# -------------------------
graph = StateGraph(WFState)
graph.add_node("generate_image", generate_image)
graph.add_node("caption_image", caption_image)
graph.add_node("build_summary", build_summary)

graph.add_edge(START, "generate_image")
graph.add_edge("generate_image", "caption_image")
graph.add_edge("caption_image", "build_summary")
graph.add_edge("build_summary", END)

workflow = graph.compile()


# -------------------------
# RUN + DISPLAY
# -------------------------
def run_workflow(prompt: str) -> WFState:
    checkpoint("Starting workflow")

    initial_state: WFState = {"prompt": prompt}

    # Stream updates to show which node executed
    for event in workflow.stream(initial_state, stream_mode="updates"):
        for node_name in event.keys():
            checkpoint(f"Node executed: `{node_name}`")

    final_state: WFState = workflow.invoke(initial_state)

    # Display image + outputs
    display(Markdown("## 🖼️ Generated Image"))
    img_path = final_state.get("image_path")
    if img_path and os.path.exists(img_path):
        display(PILImage.open(img_path))
        display(Markdown(f"Saved at: `{img_path}`"))
    else:
        display(Markdown("_Image not found._"))

    display(Markdown("## 🧾 Caption (BLIP)"))
    display(Markdown(f"`{final_state.get('caption','')}`"))

    display(Markdown("## 📝 Summary"))
    display(Markdown(final_state.get("summary","").replace("\n", "  \n")))

    return final_state


# Example
final_state = run_workflow("A peaceful rural landscape with several cows standing and grazing in an open green pasture, wide grassy land stretching into the distance, a bright sun visible in a clear blue sky, gentle rolling fields, a few scattered trees near the horizon, natural daylight, realistic colors, clear visibility, sharp focus, simple composition, no people, no buildings.")
```
